```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/LucasclFerreira/IA-matriz-de-conhecimento/main/archive/coffee.csv')
```


# **Preço Diário do Café de 2000 até 2022**

```python
dados.head()
```

```python
anos = dados['Date']
fechamento = dados['Close']
fechamento_com_datas = fechamento
fechamento_com_datas.index = anos
fig = px.line(fechamento_com_datas)
fig.show()
```

```python
num_linhas = len(fechamento.index)
tam_janela = 14
num_linhas
```

```python
# declarando matriz para receber os valores da janela deslizante
tabela = np.zeros((num_linhas - tam_janela, tam_janela + 1))
tabela
```

```python
# algoritmo que monta a matriz
for i in range(num_linhas - tam_janela):
  for j in range(tam_janela + 1):
    tabela[i][j] = round(fechamento[i + j], 2)
```

```python
# transforma a matriz em DataFrame pandas e coloca a data como índice
# a data representa a partir de qual dia a janela começa
tabela = pd.DataFrame(tabela)
tabela.index = anos[:num_linhas - tam_janela]
```

```python
tabela_sem_col_14 = tabela.drop(columns=[14])
fig_dif = px.line(tabela_sem_col_14.T.describe().loc['mean'] - tabela[14])
fig_dif.show()
```
